```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s, t = np.cos(x), np.sin(x), np.tan(x)

plt.subplot(221)
plt.plot(x, c)
plt.show()

plt.subplot(222)
plt.plot(x, s)
plt.show()

plt.subplot(223)
plt.plot(x, c)
plt.plot(x, s)
plt.show()

plt.subplot(224)
plt.plot(x, t)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s, t = np.cos(x), np.sin(x), np.tan(x)
plt.plot(x, c)
plt.plot(x, s)
ax = plt.gca()
print(ax.xaxis)
print(ax.yaxis)
plt.show()
```
